# Edge weight check

```python
assert sc_new['ibd_sum'].sum() == sc_old['ibd_sum'].sum()
assert vu_new['ibd_sum'].sum() == vu_old['ibd_sum'].sum()
assert we_new['ibd_sum'].sum() == we_old['ibd_sum'].sum()
assert nc_new['ibd_sum'].sum() == nc_old['ibd_sum'].sum()
```

# Isimorphism check

```python
G_sc_new = nx.from_pandas_edgelist(sc_new, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())
G_sc_old = nx.from_pandas_edgelist(sc_old, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())

G_vu_new = nx.from_pandas_edgelist(vu_new, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())
G_vu_old = nx.from_pandas_edgelist(vu_old, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())

G_we_new = nx.from_pandas_edgelist(we_new, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())
G_we_old = nx.from_pandas_edgelist(we_old, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())

G_nc_new = nx.from_pandas_edgelist(nc_new, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())
G_nc_old = nx.from_pandas_edgelist(nc_old, source='node_id1', target='node_id2', edge_attr=['ibd_sum'], create_using=nx.Graph())
```

```python
assert nx.vf2pp_is_isomorphic(G_sc_new, G_sc_old, node_label=None)
assert nx.vf2pp_is_isomorphic(G_vu_new, G_vu_old, node_label=None)
assert nx.vf2pp_is_isomorphic(G_we_new, G_we_old, node_label=None)
assert nx.vf2pp_is_isomorphic(G_nc_new, G_nc_old, node_label=None)
```
